```python
x = np.arange(0, 10, 1)
y = x ** 2

yerr = 0.1 + 0.2*np.sqrt(x)
xerr = 0.1 + yerr

# First illustrate basic pyplot interface, using defaults where possible.
plt.figure()
plt.errorbar(x, y, xerr= 0.5, yerr=10)
plt.title("Errorbar Plot")
```

```python
data = np.random.randn(10000)
#data = np.random.rand(1000)
#plt.hist(data);
plt.hist(data, bins=50, density=True) #, normed=True);
```

```python
x = np.random.randn(10000)
y = np.random.randn(10000)
#plt.hist2d(x, y, bins=30, cmap='Blues');
plt.hexbin(x, y, bins=30, cmap='Blues');
plt.colorbar()
```

```python
x = np.arange(0,10,0.2)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([0,0,0.9,0.9])
ax.plot(x, y, label="sine")
ax.plot(x, y + 0.5, label="Inverse of Sine")
ax.legend(loc=5) 
```

```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
plt.imshow(digits.data[10].reshape(8,8), cmap=plt.cm.gray_r)
```

```python
fig, ax = plt.subplots(8, 8, figsize=(6, 6))
for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax2 = fig.add_axes([0.39, 0.6, 0.2, 0.2])

x = np.linspace(0, 10)
ax1.plot(np.sin(x))
ax2.plot(np.sin(x))
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '--b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
leg = ax.legend(loc=0);
```
